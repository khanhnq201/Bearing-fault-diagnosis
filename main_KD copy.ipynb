```python
import cfg
from src import data_loader as data_loader, models, trainer, evaluate
from torchinfo import summary
import torch.nn as nn
import torch.optim as optim
import numpy as np
import torch

accuracy_list = []

for i in range(30):
    trainer.set_seed(i)

    X_train, Y_train, X_val, Y_val, X_test, Y_test, test_files, len_processed = data_loader.data_import(cfg, 500)
    X_train = X_train *1e2
    X_val = X_val *1e2
    X_test = X_test *1e2
    train_loader, val_loader, test_loader = data_loader.create_dataloaders(X_train, Y_train, X_val, Y_val,X_test, Y_test, cfg, len_processed)

    teacher_model = models.Deeper1DCNN(input_channels=1, num_classes=cfg.NUM_CLASSES) # Hoc model ln hn bn  dng
    teacher_checkpoint = torch.load('teacher-test.pth', map_location=cfg.DEVICE)
    teacher_model.load_state_dict(teacher_checkpoint['model_state_dict']) #what this
    
    # 2. Khi to Student model
    # Bn c th chn bt k student model no bn mun
    student_model = models.Narrow_1DCNN(input_channels=1, num_classes=cfg.NUM_CLASSES) 

    # 3. Khi to optimizer CHO STUDENT MODEL
    optimizer = optim.Adam(student_model.parameters(), lr=cfg.LEARNING_RATE)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=cfg.NUM_EPOCHS, eta_min=cfg.LEARNING_RATE * 1e-2)

    # 4. Gi hm train_distillation
    trained_model, history = trainer.train_distillation(
        teacher_model=teacher_model,
        student_model=student_model,
        train_loader=train_loader,
        val_loader=val_loader,
        optimizer=optimizer,
        temperature=cfg.TEMPERATURE,
        alpha=cfg.ALPHA,
        scheduler=scheduler,
        num_epochs=cfg.NUM_EPOCHS,
        device=cfg.DEVICE
    )
    accuracy = evaluate.plot_confusion_matrix(trained_model, test_loader, 'cpu', cfg.CLASS_NAMES)
    accuracy_list.append(accuracy)
    evaluate.plot_history(history)

print('='*20)
print('Mean accuracy:', np.mean(accuracy_list))
print('Std accuracy:', np.std(accuracy_list))
```
